```python
def show_dict_creator(show):
    base_url = 'https://www.imdb.com'
    test_dict = {}
    test_dict['title'] = show.find('a').find('img')['alt']
    for item in show.find_all('span',attrs={'class': None}):
        try:
            if item['name'] == 'ir':
                test_dict['imdb_rating'] = item['data-value']
            if item['name'] == 'nv':
                test_dict['num_imdb_ratings'] = item['data-value']
        except KeyError:
            continue
    test_dict['url'] = base_url + show.find('a')['href']
    return test_dict
```
